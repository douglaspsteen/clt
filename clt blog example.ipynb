```python
import numpy as np
import matplotlib.pyplot as plt
```

# Uniform Distribution Example

For the first example, let's consider a regular six-sided die. The probability of rolling any number, 1 through six on a given roll is equal. This is an example of a uniform distribution. However, if we collect a large number of random samples from this distribution, the distrubtion of all of the sample means will tend toward a normal distribution.

```python
die = [1,2,3,4,5,6]

plt.hist(die, bins=6, density=True)
plt.xlabel('Possible Die Roll')
plt.ylabel('Probability')
plt.title('Uniform Distribution for a Fair Six-sided Die');
```

```python
# Code to generate k = 10 random samples of size n = 5

# Container to hold the means for each sample
sample_means_10 = []

# Loop to calculate the mean of each random sample from the distribution
for i in range(10):
    sample = np.random.randint(low=1, high=7, size=5)
    sample_means_10.append(sample.mean())
```

```python
plt.hist(sample_means_10)
plt.xlabel('Mean Die Roll')
plt.ylabel('Frequency')
plt.title('k=10 Random Samples from Uniform Distribution (sample size n=5)');
```

# Binomial Distribution Example

For the second example, let's consider a magazine telemarketer making cold calls. The probability of success (making a sale) for any given call is about 5% (5 calls in 100). This is an example of the binomial distribution. However, as in the first example, if we collect a large number of random samples from this distribution, the distrubtion of all of the sample means will tend toward a normal distribution.

```python
call_result = ['Failure', 'Success']
call_prob = [0.95, 0.05]

plt.bar(x=call_result, height=call_prob)
plt.xlabel('Cold Call Result')
plt.ylabel('Probability')
plt.title('Binomial Distribution for Cold Call Success');
```

```python
# Code to generate 10 random samples of size 5

sample_means_10 = []
for i in range(10):
    sample = np.random.binomial(n=100, p=0.05, size=5)
    sample_means_10.append(sample.mean())
```

```python
plt.hist(sample_means_10)
plt.xlabel('Mean % of Successful Cold Calls')
plt.ylabel('Frequency')
plt.title('n=10 Random Samples from Binomial Distribution (sample size=5)');
```

```python
# Code to generate 100 random samples of size 5

sample_means_100 = []
for i in range(100):
    sample = np.random.binomial(n=100, p=0.05, size=5)
    sample_means_100.append(sample.mean())
```

```python
plt.hist(sample_means_100)
plt.xlabel('Mean % of Successful Cold Calls')
plt.ylabel('Frequency')
plt.title('n=100 Random Samples from Binomial Distribution (sample size=5)');
```

```python
# Code to generate 1000 random samples of size 5

sample_means_1000 = []
for i in range(1000):
    sample = np.random.binomial(n=100, p=0.05, size=5)
    sample_means_1000.append(sample.mean())
```

```python
# Code to generate 10,000 random samples of size 5

sample_means_10000 = []
for i in range(10000):
    sample = np.random.binomial(n=100, p=0.05, size=5)
    sample_means_10000.append(sample.mean())
```

# Combining Example 2 plots into one figure

```python
fig, axs = plt.subplots(2, 2, figsize=(14,10))
axs[0, 0].hist(sample_means_10)
axs[0, 0].set_title('Random Samples from Binomial Distribution (k=10, n=5)')
axs[0, 1].hist(sample_means_100)
axs[0, 1].set_title('Random Samples from Binomial Distribution (k=100, n=5)')
axs[1, 0].hist(sample_means_1000)
axs[1, 0].set_title('Random Samples from Binomial Distribution (k=1,000, n=5)')
axs[1, 1].hist(sample_means_10000)
axs[1, 1].set_title('Random Samples from Binomial Distribution (k=10,000, n=5)')

for ax in axs.flat:
    ax.set(xlabel='Mean % of Successful Cold Calls', 
           ylabel='Frequency')

# # Hide x labels and tick labels for top plots and y ticks for right plots.
# for ax in axs.flat:
#     ax.label_outer()
```
